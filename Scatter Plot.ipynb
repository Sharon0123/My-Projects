<h2 id="types_objects" align="center">Scatter Plot</h2>

```python
x=[1,2,3,4,5,6]
y=[10,50,100,80,60,90]

plt.scatter(x,y)
plt.show()
```

#### colour

```python
x=[1,2,3,4,5,6]
y=[10,50,100,80,60,90]

plt.scatter(x,y,color="red")
plt.show()
```

#### size

```python
plt.scatter(x,y,color="green",s=90)
plt.show()
```

#### alpha=> transparancy

```python
plt.scatter(x,y,color="red",s=100,alpha=0.5)
plt.show()
```

#### markers

```python
plt.scatter(x,y,color="red",s=200,alpha=0.6,marker="*")
plt.show()
```

#### edge color

```python
plt.scatter(x,y,color="violet",s=200,alpha=0.6,marker="*",edgecolors="g")
plt.show()
```

#### edge line width

```python
plt.scatter(x,y,color="violet",s=200,alpha=0.7,marker="*",edgecolors="b",linewidths=2)
plt.show()
```

### colorful scatter plot

#### colour scheme

```python
color=["red","green","blue","violet","brown","orange"]
plt.scatter(x,y,c=color)
plt.show()
```

#### size levels

```python
color=["red","green","blue","violet","brown","orange"]
size=[10,100,20,90,200,70]
plt.scatter(x,y,c=color,s=size)
plt.show()
```

#### alpha levels

```python
color=["red","green","blue","violet","brown","orange"]
size=[10,100,20,90,200,70]
al=[0.2,0.4,0.1,0.8,0.10,0.7]
plt.scatter(x,y,c=color,s=size,alpha=al)
plt.show()
```

#### color bar for color map

```python
color=["red","green","blue","violet","brown","orange"]
col=[200,100,300,700,500,400]
size=[10,100,20,90,50,70]
plt.scatter(x,y,c=col,s=size,alpha=0.7,cmap="viridis")
plt.colorbar()
plt.show()
```

#### to set name to color bar

```python
color=["red","green","blue","violet","brown","orange"]
col=[200,100,300,700,500,400]
size=[10,100,20,90,50,70]
plt.scatter(x,y,c=col,s=size,alpha=0.7,cmap="viridis")
t=plt.colorbar()
t.set_label("ColorBar")
plt.show()
```

#### color bar, legend, title, x lable, y lable, legend lable

```python
x=[1,2,3]
y=[10,50,100]
plt.scatter(x,y,color="black",label="Count")
plt.colorbar()
plt.legend()
plt.title("Scatter Plot")
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```

### x limit, y limit
#### Velocity m/s

```python
x=[0,5,10,15,20,25,30]
y1=[10,15,20,20,30,15,0]
y2=[10,12,15,12,20,10,0]
plt.scatter(x,y1,label='line-1')
plt.scatter(x,y2,label='line-2')

plt.title('Velocity-Time')
plt.xlabel("Velocity m/s")
plt.ylabel('Time(s)')
plt.xlim(0,32)
plt.ylim(0,35)
plt.legend()
plt.show()
```

### Scatter plot of Mark comparion

```python
subjects=['Tamil','English','Maths','Biology','Chemistry','Physics','Computer']
Raju=[87,82,78,73,66,75,85]
Bala=[80,70,61,78,83,87,82]
Gopi=[86,72,63,71,82,78,92]

plt.scatter(subjects,Raju,label='Raju',marker='o',color='g',edgecolors="y",linewidths=1)
plt.scatter(subjects,Bala,label='Bala',marker='o',color='r',edgecolors="y",linewidths=1)
plt.scatter(subjects,Gopi,label='Gopi',marker='o',color='b',edgecolors="y",linewidths=1)

plt.legend()
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Marks Comparison')
plt.show()
```

#### 2022 & 2023

```python
x=[1,2,3,4,5,6]

y=[10,50,100,80,60,90]
size=[10,100,20,90,50,70]
plt.scatter(x,y,s=size,alpha=1,color="yellow",label="2022")

y=[100,70,10,50,30,20]
size=[70,50,90,20,100,10]
plt.scatter(x,y,s=size,alpha=1,color="green",label="2023")

plt.legend()
plt.title("Scatter Plot",fontsize=10)
plt.show()
```
