<h2 id="types_objects" align="center">Line Plot</h2>

#### without giving X axis
##### here X axis will be considered as 0

```python
plt.plot([100,200,300,400,500])
plt.show()
```

### line plot using numpy

```python
xpts=np.array([0,4])
ypts=np.array([0,100])

plt.plot(xpts,ypts)
plt.show()
```

### linear line chart

```python
x=[1,2,3,4,5,6]
y=[10,50,100,80,60,90]
plt.plot(x,y)
plt.show()
```

#### line color

```python
plt.plot(x,y,color="green")
plt.show()
```

#### line style

```python
plt.plot(x,y,color="b",linestyle='dashed')
plt.show()
```

#### markers

```python
plt.plot(x,y,color="violet",marker="*")
plt.show()
```

#### marker size

```python
plt.plot(x,y,color="violet",marker="d",ms=20)
plt.show()
```

#### alpha=> transparancy

```python
plt.plot(x,y,color="violet",marker="d",ms=20,alpha=0.6)
plt.show()
```

#### marker edge colour (mec) (Outline)

```python
plt.plot(x,y,color="violet",marker="d",ms=20,alpha=0.6,mec="k")
plt.show()
```

#### marker face colour (mfc)

```python
plt.plot(x,y,color="violet",marker="d",ms=20,alpha=0.6,mec="k",mfc="y")
plt.show()
```

### smooth line chart using numpy

```python
x=np.linspace(0,20,100)
plt.plot(x,np.sin(x))
plt.show()
```

#### title, x lable, y lable

```python
x=[0,1,2,3,4,5]
y=[0,10,30,20,40,50]
```

```python
plt.title('Simple line plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.plot(x,y)
plt.show()
```

### line chart for square numbers

```python
x=[0,1,2,3,4,5]
y=[i**2 for i in x]
```

```python
plt.title('Simple line plot')
plt.xlabel('X-Axis')
plt.ylabel('X-Squares')
plt.plot(x,y)
plt.show()
```

### using np array

```python
x=np.arange(0,10,2)
y=[i**2 for i in x]

plt.title('Simple line plot')
plt.xlabel('X-Axis')
plt.ylabel('X-Squares')
plt.plot(x,y)
plt.show()
```

#### with markers and styles for square numbers

```python
x=np.arange(0,10,2)
y=x**2

plt.title('Simple line plot')
plt.xlabel('X-Axis')
plt.ylabel('X-Squares')
plt.plot(x,y,marker=".",ms=20,mec="k",mfc="y")
plt.show()
```

### x limit, y limit
#### Velocity m/s

```python
x=[0,5,10,15,20,25,30]
y1=[10,15,20,20,30,15,0]
y2=[10,12,15,12,20,10,0]
plt.plot(x,y1,label='line-1')
plt.plot(x,y2,label='line-2')

plt.title('Velocity-Time')
plt.xlabel("Velocity m/s")
plt.ylabel('Time(s)')
plt.xlim(0,35)
plt.ylim(0,35)
plt.legend()
plt.show()
```

### line plot of Mark comparion

```python
subjects=['Tamil','English','Maths','Biology','Chemistry','Physics','Computer']
Raju=[87,82,78,73,66,75,85]
Bala=[80,70,61,78,83,87,82]
Gopi=[86,72,63,71,82,78,92]
```

```python
plt.plot(subjects,Raju,label='Raju',marker='o',markerfacecolor='g')
plt.plot(subjects,Bala,label='Bala',marker='o',markerfacecolor='g')
plt.plot(subjects,Gopi,label='Gopi',marker='o',markerfacecolor='g')

plt.legend()
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Marks Comparison')
plt.show()
```

## previous 5 years & Present year record

```python
x=[1,2,3,4,5,6]
```

```python
# 2018
y=[50,20,60,40,50,70]
size=[20,90,30,40,50,60]
plt.plot(x,y,ms=10,color="red",linestyle='dotted',label="2018",marker=".",mec="k")

# 2019
y=[60,90,80,30,40,30]
size=[10,100,20,90,40,70]
plt.plot(x,y,ms=10,color="orange",linestyle='dotted',label="2019",marker=".",mec="k")

# 2020
y=[40,30,50,20,90,60]
size=[90,80,10,20,30,50]
plt.plot(x,y,ms=10,color="yellow",linestyle='dashdot',label="2020",marker=".",mec="k")

# 2021
y=[10,50,100,80,60,90]
size=[10,100,20,90,50,70]
plt.plot(x,y,ms=10,color="greenyellow",linestyle='dashdot',label="2021",marker=".",mec="k")

# 2022
y=[30,40,70,60,70,80]
size=[10,100,20,90,50,70]
plt.plot(x,y,ms=10,color="yellowgreen",linestyle='dashed',label="2022",marker=".",mec="k")

# 2023
y=[100,70,10,50,30,20]
size=[50,100,60,90,80,70]
plt.plot(x,y,ms=15,color="green",label="2023",marker="*",mec="b",mfc='g')

plt.legend()
plt.title("Yearly Growth",fontsize=20)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```
